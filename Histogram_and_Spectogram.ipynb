```python
import librosa
import matplotlib.pyplot as plt

# Load an audio file
filename = '/content/drive/My Drive/abc.wav'
y, sr = librosa.load(filename)

# Plot histogram of audio samples
plt.figure(figsize=(10, 6))
plt.hist(y, bins=100, color='blue', alpha=0.7)
plt.title('Histogram of Audio Samples')
plt.xlabel('Amplitude')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load an audio file
filename = '/content/drive/My Drive/abc.wav'
y, sr = librosa.load(filename)

# Compute the Short-Time Fourier Transform (STFT)
D = np.abs(librosa.stft(y))

# Convert the amplitude to decibels
DB = librosa.amplitude_to_db(D, ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(12, 8))
librosa.display.specshow(DB, sr=sr, x_axis='time', y_axis='log', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
